```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
eC = pd.read_csv (r'energyC.csv')
gC = pd.read_csv(r'gdpC.csv')
```

```python
eC
```

```python
gC
```

```python
trueDF = pd.concat([eC,gC], axis = 1)
```

```python
trueDF
```

```python
workDF = trueDF.drop(['3'], axis=1)
```

```python
workDF.columns= ['Year','Energy Production Quad BTU','GDP']
```

```python
workDF = workDF.drop([0])
```

```python
workDF
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = workDF.iloc[:,2].values.reshape(-1,1).astype(float) #GDP will be X to use predict the Y Energy Production
```

```python
X
```

```python
X.shape
```

```python
Y = workDF.iloc[:,1].values.reshape(-1,1).astype(float)
```

```python
Y
```

```python
Y.shape
```

```python
plt.scatter(X,Y)
plt.xlabel('GDP')
plt.ylabel('Energy Production Quad BTU')
```

```python
#Solid visible trendline
lr = LinearRegression()
modelLR = lr.fit(X,Y)
```

```python
yPredLR = modelLR.predict(X)
```

```python
scoreLR = modelLR.score(X,Y)
```

```python
scoreLR
```

```python
import math
rLR = math.sqrt(scoreLR)
rLR
```

```python
plt.scatter(X,Y)
plt.plot(X,yPredLR,color='red')
plt.xlabel('GDP')
plt.ylabel('Energy Production Quad BTU')
```

```python
#Training model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.25,random_state = 4444 )
```

```python
X,Y
```

```python
print('X Train: \n',X_train)
print('Y Train: \n',Y_train)
print('X Test: \n',X_test)
print('Y Test: \n',Y_test)
```

```python
modelLR = lr.fit(X_train,Y_train)
modelLR.score(X_train,Y_train)
```

```python
#98.96% 
```

```python
Y_train_pred = modelLR.predict(X_train)
```

```python
Y_train_pred, Y_train
```

```python
Y_test_pred = modelLR.predict(X_test)
Y_test_pred, Y_test
```

```python
modelLR.score(X_test,Y_test)
```
